```python
from geopy.geocoders import Nominatim
import re
import time
import csv
geolocator = Nominatim()
zipcode = []

for n in range(16110, ziptodo.shape[0]):
    location = geolocator.reverse(ziptodo["Coord"][n], timeout=50)
    if location.address == None:
        zipcode = "NaN"
    else:
        num = str(re.findall(r"\D(\d{5})\D", location.address))
        num = str(filter(str.isdigit, num))
        zipcode = num
    output = str(n)+","+ str(ziptodo["Coord"][n]) + "," + str(zipcode)
    print output
    time.sleep(1)
    with open(r"zipcode_output.csv", 'a') as f:
        writer = csv.writer(f,delimiter='\t')
        writer.writerow([output])
```

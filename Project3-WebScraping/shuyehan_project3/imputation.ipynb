```python
postcode_group=pd.read_pickle('/Users/hanshuye/Desktop/extend_info.pkl')[['area','postcode']]
postcode_group['area']=extend_information['area']
postcode_group
```

```python
# now let's concentrate on postcode
# We impute the value of postcode with the most frequent value in the group of area

area_group=postcode_group.groupby(['area']).agg(lambda x:x.value_counts().index[0])
area_group=pd.DataFrame({'area':area_group.index,'postcode':area_group['postcode']})
area_group

for index, row in postcode_group.iterrows():
    if (row['postcode']==""):
        row['postcode']=area_group[area_group['area']==row['area']]['postcode'][0]
        
postcode_group
```

```python
extend_information['postcode']=postcode_group['postcode']
extend_information
```

```python
# now lets split the price part into two columns containing the lowest price and the highest price respectively
extend_information['high']=[""]*len(extend_information)
extend_information['low']=[""]*len(extend_information)
for index, row in extend_information.iterrows():
    if (row['price']==""):
        row['low']=""
        row['high']=""
    else:
        #print row['price']
        if (row['price'].find('-')>0):
            lowindex=row['price'].index('-')
            highindex=row['price'].index('(')
            low_price=row['price'][1:lowindex-1]
            high_price=row['price'][lowindex+3:highindex-1]
            row['high']=high_price
            row['low']=low_price
        else:
            row['high']=row['price'][1:]
            row['low']=row['price'][1:]
        
extend_information
             
```

```python
# Then, we would modify the reviews column
for index, row in extend_information.iterrows():
    reviews=row['reviews'].replace(',',"")
    reviews=reviews[:reviews.index(' ')]
    row['reviews']=reviews
extend_information
```

```python
# Next we would handle the Highlight part
# Since there would be so many highlight components, we would not 
# count in all of them, but just several top frequent elements
import collections
cnt = collections.Counter()
def findOccurences(s, ch):
    return [i for i, letter in enumerate(s) if letter == ch]

for index, row in extend_information.iterrows():
    separate=findOccurences(row['highlight'],'*')
    start=0
    for i in separate:
        word=row['highlight'][start:i]
        cnt[word.lower()]+=1
        start=i+1
cnt
```

```python
# It seems that there aren't so many labels
# So we can set each of the label as a new column for the convenience of the calculation afterwards
for i in cnt.keys():
    extend_information[i]=["0"]*len(extend_information)

for index, row in extend_information.iterrows():
    row['highlight']=row['highlight'].lower()
    for j in cnt.keys():
        if (j in row['highlight']):
            row[j]="1"
            
extend_information
```
